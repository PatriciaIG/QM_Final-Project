```python
#Task1
import pandas as pd
df = pd.read_csv('coursework_1_data_2019.csv')
#Data size
print(f"Data frame is {df.shape[0]:,} x {df.shape[1]}")
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
plt.plot(df['local_authority_area'],df['2008_cases_total'],color='blue')
plt.plot(df['local_authority_area'],df['2013_cases_total'],color='green')
plt.plot(df['local_authority_area'],df['2018_cases_total'],color='red')
plt.ylim(0,2000)
plt.show()
```

```python
#Places where were reported less cases of childhood obessity
df[df['2018_cases_total']<=df['2008_cases_total']]['local_authority_area']
```

```python
#Percentaje childhood obesity cases per hundred citizens %
#Childhood obesity cases/ hundred population 
ratio_child_cit=['ratio_child_cit_2008','ratio_child_cit_2013','ratio_child_cit_2018']
df_var_rate=df.iloc[:,0:1]
for b in ratio_child_cit:
    df_var_rate[b]=df[f"{b[-4:]}_cases_total"]/df[f"{b[-4:]}_pop_total"]*100
    print(f"Percentaje of obese child per citizen in {b[-4:]}")
```

```python
df_var_rate.head()
```

```python
df_var_rate['var_rate']=df_var_rate['ratio_child_cit_2018']-df_var_rate['ratio_child_cit_2008']
```

```python
#Places where were reported less rates of childhood obessity
df[df_var_rate['ratio_child_cit_2018']<=df_var_rate['ratio_child_cit_2008']]['local_authority_area']
```

```python
##Pobacion promedio
df['aver_pop']=round((df['2008_pop_total']+df['2013_pop_total']+df['2018_pop_total'])/3)
```

```python
list(df.columns.values)
```

```python
#Dataframe de tasas de aumento de obesidad seg'un inversion--> Analisis 1
df_a1=pd.concat([df_var_rate.iloc[:,0],df_var_rate.iloc[:,4],df.iloc[:,19],df.iloc[:,28],df.iloc[:,27]],axis=1)
```

```python
df_a1.head()
```

```python
##dummies variables para local authority
df_dummies=pd.get_dummies(df_a1.iloc[:,4])
df_dummies.head()
df_dummies=df_dummies.drop(columns=['other_london'])
```

```python
df_analisis=pd.concat([df_a1.iloc[:,0:4],df_dummies.iloc[:,:]],axis=1)
```

```python
df_analisis.head()
```

```python
df_analisis_final=df_analisis.drop(columns=['local_authority_area'])
```

```python
df_analisis_final.head()
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

import numpy as np

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sn

pd.set_option('display.max_rows', 300) # specifies number of rows to show
pd.options.display.float_format = '{:40,.4f}'.format # specifies default number format to 4 decimal places
plt.style.use('ggplot') # specifies that graphs should use ggplot styling
%matplotlib inline

df = df_analisis_final
plt.rcParams["axes.grid"] = False
f = plt.figure(figsize=(19, 15))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=14)
plt.title('Correlation Matrix', fontsize=16)
```

```python
model_var_rate = sm.formula.ols('var_rate ~ total_budget + aver_pop + london_borough + metropolitan_borough + non_metropolitan_county + unitary_authority', df_analisis_final).fit()
model_var_rate.summary()
```

```python
#VIF values calcules
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
y, X = dmatrices('var_rate ~ total_budget + aver_pop + london_borough + metropolitan_borough + non_metropolitan_county + unitary_authority', data=df, return_type='dataframe')
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns
vif
```
