```python
import pandas as pd
from sklearn.linear_model import LinearRegression
import statsmodels.api as sm

import numpy as np

import sklearn
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix

import matplotlib.pyplot as plt
import seaborn as sn

pd.set_option('display.max_rows', 300) # specifies number of rows to show
pd.options.display.float_format = '{:40,.4f}'.format # specifies default number format to 4 decimal places
plt.style.use('ggplot') # specifies that graphs should use ggplot styling
%matplotlib inline

plt.rcParams["axes.grid"] = False
f = plt.figure(figsize=(10, 18))
plt.matshow(df.corr(), fignum=f.number)
plt.xticks(range(df.shape[1]), df.columns, fontsize=14, rotation=45)
plt.yticks(range(df.shape[1]), df.columns, fontsize=14)
cb = plt.colorbar()
cb.ax.tick_params(labelsize=16)
plt.title('Correlation Matrix', fontsize=10)
```

```python
model = sm.formula.ols('median_house_price ~perc_employment_rate+perc_professional_working_age', df).fit()
model.summary()
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
y, X = dmatrices('median_house_price ~perc_employment_rate+perc_professional_working_age', data=df, return_type='dataframe')
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns
vif
```

```python
model_var_rate = sm.formula.ols('ratio_evsp ~ own_outright + mortgage + rented_pl+work_population+born_abroad', df).fit()
model_var_rate.summary()
```

```python
from patsy import dmatrices
from statsmodels.stats.outliers_influence import variance_inflation_factor
y, X = dmatrices('ratio_evsp ~ own_outright + mortgage + rented_pl+work_population+born_abroad', data=df, return_type='dataframe')
vif = pd.DataFrame()
vif['VIF'] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
vif['variable'] = X.columns
vif
```

```python
import matplotlib.pyplot as plt
plt.scatter(model_var_rate.fittedvalues, model_var_rate.resid)
# adding title and labels
plt.xlabel('Fitted ratio_evsp')
plt.ylabel('Residual')
plt.title('Residual vs. Fitted ratio_evsp')
plt.show()
```
