```python
from scipy import stats
independent_variables=list(df.drop(columns=['Codes', 'Names']).columns)
variable=['% Good health','% employment','% No qualifications','% Activities limited','% No cars in household']
D_val=[]
p_val=[]
for i in independent_variables:
    d,p=stats.kstest(df[i], 'norm')
    D_val.append(d)
    p_val.append(p)
data={'Variable':variable,'D':D_val,'p-value':p_val} 
norm_test=pd.DataFrame(data)
pd.options.display.float_format = '{:.3f}'.format
display(norm_test)
```
